### Ex E5 ⭐⭐ — Fonction extract()
Encapsulez tout dans une fonction `extract(q1_path, q2_path, clients_path, produits_path) -> dict`
qui retourne un dictionnaire avec les 4 DataFrames harmonisés et consolidés.
La fonction doit afficher des logs à chaque étape.

```python
import pandas as pd
import logging as lg

lg.basicConfig(level=lg.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def import_dataframe_from_csv(name, df_path):
    lg.info(f"Import des données de : {name}")
    df = pd.read_csv(df_path)
    if not df.empty:
        lg.info(f"Import des données de {name} réussi")
        lg.info(f"{df.shape}")
    else:
        lg.error(f"Import des données de {name} échoué")
        return None
    return df

def harmoniser_commandes(dfq1, dfq2):
    lg.info("Harmonisation des données de commandes")
    lg.info("Ajout de la source")
    dfq1['source'] = 'Q1_2024'
    dfq2['source'] = 'Q2_2024'
    lg.info("Harmonisation des colonnes")
    dfq2.rename(columns={'moyen_paiement': 'mode_paiement'}, inplace=True) #renommer une colonne
    dfq1["devise"] = "EUR"
    return True

def extract(q1_path, q2_path, clients_path, produits_path):
    lg.info('Import des données')
    df_cmdq1 = import_dataframe_from_csv('df_cmdq1', q1_path)
    df_cmdq2 = import_dataframe_from_csv('df_cmdq2', q2_path)
    df_clients = import_dataframe_from_csv('df_clients', clients_path)
    df_produits = import_dataframe_from_csv('df_produits', produits_path)

    if harmoniser_commandes(df_cmdq1, df_cmdq2):
        lg.info("Harmonisation des données de commandes réussie")
        df_conso = pd.concat([df_cmdq1, df_cmdq2])
        lg.info("Concaténation des DataFrame Commandes effectuée")
    else:
        lg.error("Echec sur l'harmonisation des commandes")
        return None
    
    dico_df = {"df_cmd": df_conso,"df_clients": df_clients,"df_produits": df_produits}
    if len(dico_df) == 3:
        lg.info("Import Réussi")
    else:
        lg.error("Echec de l'import")
        return None
    
    return dico_df

dict_df = extract("data/commandes_q1.csv","data/commandes_q2.csv","data/clients.csv","data/produits.csv")
```

### Ex T3 ⭐ — Corriger les types
1. Convertissez `date_commande` en datetime en gérant les formats incohérents
2. Extrayez : `annee`, `mois`, `trimestre`, `jour_semaine`
3. Convertissez `quantite` en numérique avec gestion des erreurs
4. Affichez les types avant et après

```python
#Conversion des type quantite en numerique, si erreur, il va lever une exception (raise)
commandes['quantite'] = pd.to_numeric(commandes['quantite'], errors='raise')

#Conversion des Dates en datetime, le format ISO permet de transfomer auant les YEAR-MONTH-DAY en datetime que les YEAR/MONTH/DAY
commandes['date_commande'] = pd.to_datetime(commandes['date_commande'], errors='raise', format='ISO8601')

#Extraction des données de la date
def conversion_date(date):
    annee = date.year
    mois = date.month
    trimestre = (date.month - 1) // 3 + 1
    jour_semaine = date.day_name(locale='fr_FR')
    return {'a':annee, 'm':mois, 't':trimestre, 'j':jour_semaine}

#Création des nouvelles colonnes dérivées de la date
commandes['annee'] = commandes['date_commande'].apply(lambda x: conversion_date(x)['a'])
commandes['mois'] = commandes['date_commande'].apply(lambda x: conversion_date(x)['m'])
commandes['trimestre'] = commandes['date_commande'].apply(lambda x: conversion_date(x)['t'])
commandes['jour_semaine'] = commandes['date_commande'].apply(lambda x: conversion_date(x)['j'])

commandes.dtypes
```

### Ex T8 ⭐⭐⭐ — Validation qualité
Écris une fonction `valider_pipeline(df) -> bool` qui :
1. Vérifie qu'il n'y a plus de NaN sur les colonnes critiques
2. Vérifie qu'il n'y a plus de doublons
3. Vérifie que tous les prix sont > 0
4. Vérifie que toutes les quantités sont > 0
5. Vérifie que montant_total == quantite × prix_unitaire (à 1 centime près)
6. Retourne True si tout est OK, False sinon avec les erreurs détaillées

```python
def valider_pipeline(df: pd.DataFrame) -> bool:
    '''Fonction de validation du pipeline de données de ventes'''
    '''Vérification effectuées : NaN, Doublons, Prix > 0, Quantites > 0, Montant total == quantite * prix_unitaire '''
    '''Retourne True si tout est OK, False sinon avec les erreurs détaillées'''

    lg.info(f"Validation du pipeline de données de ventes")

    validation = True

    # Vérification que c'est bien un DataFrame
    if type(df) != pd.DataFrame:
        lg.error(f"Le paramètre df doit être un DataFrame")
        # Pas besoin d'aller plus loin dans ce cas
        return False
    
    # Vérification que les colonnes testés du DataFrame existe bien
    if not set(['prix_unitaire', 'quantite', 'montant_total']).issubset(df.columns):
        lg.error(f"Le DataFrame doit contenir les colonnes prix_unitaire, quantite et montant_total")
        # Pas besoin d'aller plus loin dans ce cas
        return False

    # Vérification des NaN sur les colonnes critiques
    for col in ['prix_unitaire', 'quantite', 'montant_total']:
        if df[col].isna().sum() > 0:
            lg.error(f"Il y a des NaN dans le DataFrame sur la colonne {col}")
            validation = False
    # Vérification des doublons
    if df.duplicated().sum().sum() > 0:
        lg.error(f"Il y a des doublons dans le DataFrame")
        lg.error(f"Les colonnes avec des doublons sont :\n {df.columns[df.duplicated().any()]}")
        validation = False
    # Vérification des prix > 0
    if (df['prix_unitaire'] <= 0).sum() > 0:
        lg.error(f"Il y a des prix négatif ou nul dans le DataFrame")
        lg.error(f"Les colonnes avec des prix négatifs sont :\n {df.columns[df['prix_unitaire']]}")
        validation = False
    # Vérification des quantités > 0
    if (df['quantite'] <= 0).sum() > 0:
        lg.error(f"Il y a des quantités négatives ou nulle dans le DataFrame")
        lg.error(f"Les colonnes avec des quantités négatives sont :\n {df.columns[df['quantite']]}")
        validation = False
    # Vérification des montant_total
    # commandes[commandes.montant_total != (commandes.quantite * commandes.prix_unitaire).round(2)].shape[0]
    if (df['montant_total'] != (df['quantite'] * df['prix_unitaire']).round(2)).sum() > 0:
        lg.error(f"Il y a des incohérences de montant total")
        lg.error(f"Les colonnes avec des incohérences sont :\n {df.columns[df['montant_total']]}")
        validation = False

    if validation:
        lg.info(f"Le DataFrame est valide")
    # Si toute les vérifications sont bonnes, on valide le DataFrame
    return validation

valider_pipeline(df_condense)
```
